```python
import requests


url = f"https://turkiyeapi.dev/api/v1/provinces"
try:
        response = requests.get(url)
        response.raise_for_status()  # Raise error for bad responses (4xx or 5xx)
        data = response.json()
        data
except requests.exceptions.RequestException as e:
        f"An error occurred: {e}"
```

```python
df=data
df
```

```python
# prompt: now give me a dataframe from the data above please
import pandas as pd
df = pd.DataFrame(data['data'])
df
```

```python
# Extract latitude and longitude into separate columns
df['latitude'] = df['coordinates'].apply(lambda x: x['latitude'])
df['longitude'] = df['coordinates'].apply(lambda x: x['longitude'])

# Extract region names in English and Turkish
df['region_en'] = df['region'].apply(lambda x: x['en'])


# Drop original nested columns if you want
df = df.drop(columns=['coordinates', 'region'])

df
```

```python
# Explode districts list into separate rows with province info repeated
df_districts = df[['id', 'name', 'districts']].explode('districts')

# Normalize districts dictionaries into columns
df_districts = pd.concat([
    df_districts.drop(columns=['districts']),
    df_districts['districts'].apply(pd.Series)
], axis=1)

df_districts
```

```python
# prompt: change the name column for both dataframes above to provine_name

df = df.rename(columns={'name': 'province_name'})
df_districts = df_districts.rename(columns={'name': 'province_name'})

df
df_districts
```

```python
import requests
import pandas as pd

# Define the date range for the query
start_date = "2020-09-14 10:00:00"
end_date = "2024-09-16 10:00:00"

# API URL (adjusted to encode spaces properly)
url = "https://deprem.afad.gov.tr/apiv2/event/filter"
params = {
    "start": start_date,
    "end": end_date
}

# Send GET request to AFAD API
response = requests.get(url, params=params)
data = response.json()


```

```python

# Convert to pandas DataFrame
df = pd.json_normalize(data)

df
```
